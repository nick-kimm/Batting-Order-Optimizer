```python
L = pd.read_csv('Left.csv')
R = pd.read_csv('Right.csv')
L_b = pd.read_csv('Left_b.csv')
R_b = pd.read_csv('Right_b.csv')

B = pd.read_csv('Basic.csv')


L.columns, R.columns, B.columns, L_b.columns
```

```python
B = B.loc[:, ['playerId','Name', 'SB']]
L = L.loc[:,['playerId','AVG','OBP','SLG','wRC','wOBA']]
R = R.loc[:,['playerId','AVG','OBP','SLG','wRC','wOBA']]
L_b = L_b.loc[:,['playerId','AB']]
R_b = R_b.loc[:,['playerId','AB']]
```

```python
pd.merge(left=B,right=L,left_on='playerId',right_on='playerId')
```

```python
df = pd.merge(left = B, right = L_b, left_on = 'playerId', right_on = 'playerId')
df.rename(columns = {'AB':'AB_L'}, inplace = True)
df = pd.merge(left = df, right = L,left_on = 'playerId',right_on = 'playerId')
df.rename(columns = {'AVG':'AVG_L','OBP':'OBP_L','SLG':'SLG_L','wRC':'wRC_L','wOBA':'wOBA_L'}, inplace = True)
df = pd.merge(left = df, right = R_b, left_on = 'playerId', right_on = 'playerId')
df.rename(columns = {'AB':'AB_R'}, inplace = True)
df = pd.merge(left = df, right = R,left_on = 'playerId',right_on = 'playerId')
df.rename(columns = {'AVG':'AVG_R','OBP':'OBP_R','SLG':'SLG_R','wRC':'wRC_R','wOBA':'wOBA_R'}, inplace = True)
df
```

```python
df.to_csv('MLBSplits.csv')
```
